```python
points_for_cloud=1000
num_diag_per_class=15
#r_values=[3, 3.5, 4]
r_values=[2.5,3.5,4,4.1,4.3]
labels = []
points=[]

for r in r_values:
    for _ in range(num_diag_per_class):
        
        X = np.empty([points_for_cloud, 2])
        x, y = np.random.uniform(), np.random.uniform()
        
        for i in range(points_for_cloud):
            
            X[i, :] = [x, y]
            x = (X[i, 0] + r * X[i, 1] * (1 - X[i, 1])) % 1
            y = (X[i, 1] + r * x * (1 - x)) % 1
            
        points.append(X)
        labels.append([str(r)])


   
```

```python
combined_data = list(zip(points, labels))
np.random.shuffle(combined_data)
points, labels = zip(*combined_data)
```
